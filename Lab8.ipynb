```
X = 5.85 - 0.09*8
x = [1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7]
y = [10.517, 10.193, 9.807, 9.384, 8.977, 8.637, 8.442, 8.482, 8.862, 9.707, 11.132, 13.302]
n = len(y) - 1
h = x[1] - x[0]
t = round(((X - x[0]) / h), 3)
```

Интерполяционная формула Лагранжа

```
print(f'1/{h}*(', end = '')
for i in range(n+1):
    print(f'{y[i]}*(-1)^{n-i} / ({i}!*{n-i}!)* [', end = '')
    for j in range(n+1):
        if j == i:
            continue
        print(f'(t-{j})', end = '')
    print(']\'+')
print(')', end = '')
```

```
print(f'1/{h}*(', end = '')
for i in range(n+1):
    print(f'{y[i]}*(-1)^{n-i} / ({i}!*{n-i}!)* (', end = '')
    for j in range(n+1):
        if j == i:
            continue
        for k in range(n+1):
            if k == j or k == i:
                continue
            print(f'(t-{k})', end = '')
        print('+', end = '')
    print(')+')
print(')', end = '')
```

```
#подстановка значений
print(f'1/{h}*(', end = '')
for i in range(n+1):
    print(f'{y[i]}*(-1)^{(n-i) % 2} / ({i}!*{n-i}!)* (', end = '')
    for j in range(n+1):
        if j == i:
            continue
        for k in range(n+1):
            if k == j or k == i:
                continue
            print(f'({t}-{k})', end = '')
        print('+', end = '')
    print(')+')
print(')')
```

Интерполяционная формула Ньютона

```
#tabel delta y
import copy
d = []
#d.append(y)
dn = copy.copy(y)
l = []
for i in range(n, 0, -1):
    print(n - i, end = ":  ")
    for j in range(i):
        l.append(float((float(dn[j+1]) - float(dn[j]))))
        #"{:.16f}".format(float(s))
    print(l)
    print()
    d.append(l[:])
    dn = copy.copy(l)
    l.clear()

h = round((x[1] - x[0]), 3)
print(d)
```

```
print(f'1/{h} (', end = '')

for i in range(n+1):
    if i != 0:
        print(f'{d[i-1][0]}/{i}!', end = '')
    print('(', end = '')
    for j in range(i):
        for k in range(i):
            if k == j:
                continue
            print(f'({t} - {k})', end = '')
        print('+', end  = '')

    print(')')
print(')')
```
