#Дано :  $ln(lg(x)) - \frac {1}{1+x^2} = 0$
---
# производная : $ \frac {2x}{(x^2+1)^2} + \frac {1}{xln(x)ln(10)}$


```
def F(x):
    return math.log10(math.log(x)) - 1 / (1 + x ** 2)
```

```
import math
import matplotlib.pyplot as plt
import numpy as np

plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
x = np.linspace(2, 4, 50)
y = [F(i) for i in x]
plt.xlabel("x")         # ось абсцисс
plt.ylabel("y")    # ось ординат
plt.grid(color = 'black',linewidth = 1, linestyle = '--')# включение отображение сетки
plt.axhline(np.heaviside(0, 0))
plt.plot(x, y, 'red')
```

```
#деление пополам
a = 3.25
b = 3.50
c = (a + b) / 2
while (abs(F(c)) > 0.00001):
    if (F(c)*F(a) > 0):
        a = c
    if (F(c)*F(b) > 0):
        b = c
    c = (a + b) / 2
    print(c)
```

```
#хорд
a = 3.25
b = 3.50
c = a - ((b - a) *F(a)) / (F(b) - F(a))
while (abs(F(c)) > 0.00001):
    if (F(c)*F(a) > 0):
        a = c
    if (F(c)*F(b) > 0):
        b = c
    c = a - ((b - a) *F(a)) / (F(b) - F(a))
    print(c)
```

```
#Ньютона
c = 3
while (abs(F(c)) > 0.00001):
    f = (1 / (math.log(10)*math.log(c)*c) + (2*c) / (c ** 2 + 1) ** 2)
    c = c - F(c) / f
    print(c)
```

```
#секущих
cl = 3 
c = 3.1
while (abs(F(c)) > 0.00001):
    cn = c - ((c - cl)/(F(c) - F(cl))) * F(c)
    cl = c
    c = cn
    print(cn)
```

```
#простые итераций
c = 3
while (abs(F(c)) > 0.00001):
    c = c - 6*F(c)
    print(c)
```
