```
n = 6
print("Введите x -ы:")
x = [float (i) for i in input().split()]
print("Введите y(x) -и:")
y = [float (i) for i in input().split()]
print(x)
print(y)
```

```
dw = list()
print("w(x) =", end =  " ")
for i in range(n):
    print(f"(x - {x[i]})", end = " * ")
print("\n")

print("w'(x) =", end =  " ")
for i in range(n):
    for j in range(n):
        if j == i:
            continue
        else:
            print(f"(x - {x[j]})", end = " ")
    print(" + ")
print("\n")

for i in range(n):
    print(f"w'({x[i]}) =", end =  " ")
    s = 1
    for j in range(n):
        if j == i:
            continue
        else:
            print(f"({x[i]} - {x[j]})", end = " * ")
            s *= (x[i] - x[j])
    print(f"= {round(s,6)}")
    dw.append("{:.16f}".format(float(s)))
```

```
print("L (x) = ")
for i in range(n):
    print("+", y[i], end = " ")
    for j in range(n):
        print(f"(x - {x[j]})", end = " ")
    print(" / ", end = "")
    print(f"((x - {x[i]})  ({dw[i]}))", end = "\n")
```
