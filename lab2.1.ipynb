```
a = np.copy(array)

arr = np.array([])
# print(arr)
index.clear()
# string = []
# print(string)
```

```
print(a)
flag = True
while len(a) > 0 and flag:
    aMax = a[0][0]
    for i in range(len(a)):
        for j in range(len(a)):
            if aMax <= a[i][j]:
                aMax = a[i][j]
                p = i
                q = j
                
    print("коэффициенты максимального элемента:",p, q)
    if aMax == 0:
        flag = False 
        print("Ошибка в матрице!")
        break
    print()

    for i in range(len(a)):
        m = a[i][q] / a[p][q]
        for j in range(len(a)+1):
            if i == p:
                break
            a[i][j] = a[i][j] - a[p][j] * m
        
    print("Матрица:")
    print(a, "\n")
    string = np.array([])
    
    # for i in range(n+1-len(a[p])):
    #     string.append(0) 
    string = np.append(string, a[p])
    for i in index[::-1]:
        string = np.insert(string, i, 0)
    index.append(q)

    arr = np.append(arr, np.array(string))
    #np.append(x, a[p])
    a = np.delete(a, p, 0)
    a = np.delete(a, q, 1)

    print(a, "\n")

if flag:    
    arr.shape = (3, 4)
    print("arr :",arr, sep = "\n")
    print("index :",index)
```
