ДАНО:

$\int_{0}^{1} cosh(x^2) \,dx$

image.png

```
import numpy as np
def F(x):
    return np.cosh(x*x)
a, b, n = 0, 1, 10
h = (b-a)/n
```

```
x = [round(a+h*i, 3) for i in range(n+1)]
y = [round(F(i), 5) for i in x]
mid = [round(F(i+h/2), 5) for i in x]

#Формулы прямоугольников
print("формула левых прямоугольников: ", h * sum(y[:-1]))
print("формула правых прямоугольников: ", h * sum(y[1:]))
print("формула средних прямоугольников", h * sum(mid[:-1]))

#Формула трапеций
print("формула трапеций: ", h * ((y[0] + y[-1]) / 2 + sum(y[1:-1])))

#Формула Симпсона
print("формула Симпсона: ", h/3*(-y[0]-y[-1]+ 2 * sum([y[i] for i in range(n+1) if i % 2 == 0]) + 4 * sum([y[i] for i in range(n+1) if i % 2 == 1])))
```
