# Дано:

$2y - cos(x+1) = 0$

$x + sin(y) = -0.4$

```
import numpy as np
import math
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

x = np.linspace(-1.5, 0, 50)
y = [ 0.5 * math.cos(i+1) for i in x]
plt.plot(x,y)

y = np.linspace(-0.5, 1, 50)
x = [-0.4 - math.sin(i) for i in y]
plt.plot(x,y)

# ax.legend();
plt.xlabel("x")    # ось абсцисс
plt.ylabel("y")    # ось ординат
plt.grid(color = 'black',linewidth = 1, linestyle = '--')# включение отображение сетки
plt.axhline(np.heaviside(0, 0), color = 'black')
```

# Дано:

$sin(x+y) - 1.5x = 0.1$

$x^2 + y^2 =   1 $


---


$F_1 = sin(x+y) - 1.5x - 0.1$

$F_2 = x^2 + y^2 - 1$



image.png

```
#Метод Ньютона

x, y = 1, 1
dx, dy = 1, 1
m0 = [0, 0] 
m1 = [0, 0]
v = [0, 0]
while(max(abs(dx), abs(dy)) > 0.00001):
    m0[0], m0[1] = math.cos(x+y)-1.5, math.cos(x+y)
    m1[0], m1[1] = 2*x, 2*y
    v[0] = math.sin(x+y)-1.5*x-0.1
    v[1] = x**2+y**2-1
    m = np.array([m0, m1])
    v = np.array(v)
    print(f"x = {x}     y = {y}")
    d = np.linalg.solve(m, v)
    x, y = x - d[0], y - d[1]
    dx, dy = d[0], d[1]

```
